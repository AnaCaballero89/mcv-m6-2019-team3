```python
import os

first_val_frame = 536
```

```python
#Â Crear carpetas
images_train_path = "../datasets/cars/images/train/"
images_val_path = "../datasets/cars/images/val/"
labels_train_path = "../datasets/cars/labels/train/"
labels_val_path = "../datasets/cars/labels/val/"
```

```python
if not os.path.exists(images_train_path):
    os.makedirs(images_train_path)
    
if not os.path.exists(images_val_path):
    os.makedirs(images_val_path)
    
if not os.path.exists(labels_train_path):
    os.makedirs(labels_train_path)
    
if not os.path.exists(labels_val_path):
    os.makedirs(labels_val_path)
```

```python
import glob

new_path = "data/coco/images/"

with open("../datasets/cars/train5k.txt","w") as f:
    for image_path in glob.glob (images_train_path + "*.jpg"):
        image_path = image_path.replace("../datasets/cars/images/",new_path)
        f.write(image_path + "\n")

with open("../datasets/cars/val5k.txt","w") as f:
    for image_path in glob.glob (images_val_path + "*.jpg"):
        image_path = image_path.replace("../datasets/cars/images/",new_path)
        f.write(image_path + "\n")
        
```

```python
from utils.reading import read_annotations_from_xml

gt_list, _ = read_annotations_from_xml("../annotations/m6-full_annotation.xml", "../datasets/AICity_data/train/S03/c010/vdo.avi")
```

```python
print(gt_list)
```

```python
if not os.path.exists(labels_train_path):
    os.makedirs(labels_train_path)
    
if not os.path.exists(labels_val_path):
    os.makedirs(labels_val_path)

for annotation in gt_list:
    if annotation.frame < first_val_frame:
        path = labels_train_path + "frame_{:04}".format(annotation.frame+1) + ".txt"
    else:
        path = labels_val_path + "frame_{:04}".format(annotation.frame+1) + ".txt"

    with open(path,"a") as f:
        f.write("2 {:.6f} {:.6f} {:.6f} {:.6f}\n".format(annotation.xtl/1920.0, annotation.ytl/1080.0, annotation.width/1920.0, annotation.height/1080.0))

```
